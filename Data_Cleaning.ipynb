# Data Cleaning - Superstore Sales Data

* Importing Required Libraries 

```python
import pandas as pd
```

* Load the file 

```python
df = pd.read_excel("Superstore.xls")
```

* Reviewing Data Structure

```python
df.head(5)
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

* Dropping Unwanted Columns

```python
drop = [
    'Row ID', 'Customer Name', 'Ship Mode', 'Postal Code', 'Product Name'
    ]

df.drop(columns=drop, inplace=True)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

* Finding Duplicate values

```python
duplicates = df[df.duplicated()]
print(f"Number of duplicated rows: {duplicates.shape[0]}")
duplicates
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

* Converting Date Columns 

```python
df['order_date'] = pd.to_datetime(df['Order Date'])
df['ship_date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

* Extracting Year-Month Column

```python
df['Year-Month'] = df['order_date'].dt.to_period('M').astype(str)
```

* Saving Cleaned Dataset

```python
df.to_csv('cleaned_superstore_data.csv', index=False)
```
